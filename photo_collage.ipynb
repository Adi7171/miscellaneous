## To add overlay

```python
def image_overlay(baseImg,centerImg,location):
  h,w = baseImg.shape[:2] #Not selecting the layers that is 3 in (430,430,3)
  h1,w1 = centerImg.shape[:2]
  x,y= location
  baseImg[y:y+h1,x:x+w1] = centerImg # rows height and columns is width
  return baseImg
```

## To find which location to place

```python
x= full.shape[0]//2 - c.shape[0]//2
y= full.shape[1]//2 - c.shape[1]//2
```
